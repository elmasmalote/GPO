```python
#Datos
#filas son las maquinas (de 1 a 5), columnas son las tareas (de 1 a 4)
n=4 #tareas
m=5 #maquinas
tiempos = [[5,9,9,4],
           [9,3,4,8],
           [8,10,5,8],
           [10,1,8,7],
           [1,8,6,2]]
#tiempos = [[24, 18, 21, 12, 28, 19],
 #          [23, 22,  4, 17, 24, 25]]
```

```python
#Para calcular lista de la tareas en orden crítico a colocar.
T = [sum(tiempos[j][i] for j in range(m)) for i in range(n)]
T_orden = sorted(T, reverse=True)
orden_tareas = [(T.index(T_orden[i])+1) for i in range(n)]
orden_tareas
```

```python
#Función para calcular el makespan de una determinada secuencia
def calculoMakespan(orden):
    fin =[[0 for col in range(len(orden))] for row in range(m)] #matriz de los tiempos finales de cada tarea de la secuencia en cada maquina

    #tiempos finales solo para maquina 1
    for l in range(len(orden)):
        if(l==0): fin[0][l] = tiempos[0][orden[l]-1]
        else: fin[0][l] = fin[0][l-1]+tiempos[0][orden[l]-1]   

    #tiempos finales para el resto de las maquinas
    for k in range(1,m):
        fin[k][0]=fin[k-1][0]+tiempos[k][orden[0]-1]
        for l in range(1,len(orden)):
            fin[k][l]=max(fin[k][l-1],fin[k-1][l]) + tiempos[k][orden[l]-1]

    makespan = fin[m-1][len(orden)-1]
    #print(fin)
    #print(makespan)
    return orden,makespan
```

```python
#Ver cuál es mejor secuencia entre las dos primeras tareas de la lista orden_tareas
orden1, mak1 = calculoMakespan([orden_tareas[0],orden_tareas[1]])  #[1,3]
orden2, mak2 = calculoMakespan([orden_tareas[1],orden_tareas[0]])  #[3,1]
if(min(mak1,mak2)==mak1): orden = orden1
else: orden = orden2
print(orden)

#Incorporar las siguientes tareas
secuencias = []
for i in range(2,n):  #siguientes tareas a agregar (desde posicion 2 de lista orden_tareas)
    tarea = orden_tareas[i]
    print("Tarea a incorporar: ",tarea)
    ordenes = []
    makespan = []
    for j in range(i+1):
        sec = orden.copy()
        sec.insert(j,tarea)
        orden_j, mak_j = calculoMakespan(sec)
        ordenes.append(orden_j)
        makespan.append(mak_j)
    print("Secuencias posibles: ",ordenes) 
    print("Makespans correspondientes: ",makespan)
    orden = ordenes[makespan.index(min(makespan))]
    print("Secuencia con mínimo makespan: ", orden, "\n")

print("Se ha terminado el algoritmo. \nLa solución encontrada es la secuencia: ",orden," con un makespan de ", min(makespan))
```
